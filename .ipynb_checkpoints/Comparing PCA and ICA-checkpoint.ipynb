```python
from sklearn.datasets import fetch_lfw_people
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import ConfusionMatrixDisplay


lfw_people = fetch_lfw_people(min_faces_per_person=70, resize=0.4)
X_train, X_test, y_train, y_test = train_test_split(lfw_people.data, lfw_people.target, test_size=0.25, random_state=42)

# Apply PCA to extract eigenfaces
n_components = 150
pca = PCA(n_components=n_components, whiten=True)
pca.fit(X_train)
eigenfaces = pca.components_.reshape((n_components, lfw_people.images.shape[1], lfw_people.images.shape[2]))
fig, axes = plt.subplots(10, 10, figsize=(15, 15))
for i, ax in enumerate(axes.flat):
    ax.imshow(eigenfaces[i], cmap='gray')
    ax.axis('off')
    ax.set_title(f'PCA: {i+1}')
plt.show()
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
clf = SVC(kernel='linear', C=1)
clf.fit(X_train_pca, y_train)
y_pred = clf.predict(X_test_pca)
classification_report(y_test, y_pred, target_names=target_names)
ConfusionMatrixDisplay.from_estimator(
    clf, X_test_pca, y_test, display_labels=target_names, xticks_rotation="vertical")
plt.tight_layout()
plt.show()
accuracy = accuracy_score(y_test, y_pred)
print(f'PCA accuracy: {accuracy:.3f}')
for n_components in [10, 25, 50, 75, 100]:
    pca = PCA(n_components=n_components, whiten=True)
    pca.fit(X_train)
    X_train_pca = pca.transform(X_train)
    X_test_pca = pca.transform(X_test)
    clf = SVC(kernel='linear', C=1)
    clf.fit(X_train_pca, y_train)
    y_pred = clf.predict(X_test_pca)
    accuracy = accuracy_score(y_test, y_pred)
    print(f'n_components = {n_components}, PCA accuracy: {accuracy:.3f}')


```

```python
from sklearn.datasets import fetch_lfw_people
from sklearn.decomposition import FastICA
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import ConfusionMatrixDisplay

lfw_people = fetch_lfw_people(min_faces_per_person=70, resize=0.4)
X_train, X_test, y_train, y_test = train_test_split(lfw_people.data, lfw_people.target, test_size=0.25, random_state=42)
n_components = 150
ica = FastICA(n_components=n_components, whiten='arbitrary-variance', max_iter=1000)
ica.fit(X_train)
independent_components = ica.components_.reshape((n_components, lfw_people.images.shape[1], lfw_people.images.shape[2]))
fig, axes = plt.subplots(10, 10, figsize=(15, 15))
for i, ax in enumerate(axes.flat):
    ax.imshow(independent_components[i], cmap='gray')
    ax.axis('off')
    ax.set_title(f'ICA: {i+1}')
plt.show()
X_train_ica = ica.transform(X_train)
X_test_ica = ica.transform(X_test)
clf = SVC(kernel='linear', C=1)
clf.fit(X_train_ica, y_train)
y_pred = clf.predict(X_test_ica)
classification_report(y_test, y_pred, target_names=target_names)
ConfusionMatrixDisplay.from_estimator(
    clf, X_test_ica, y_test, display_labels=target_names, xticks_rotation="vertical")
plt.tight_layout()
plt.show()
accuracy = accuracy_score(y_test, y_pred)
print(f'ICA accuracy: {accuracy:.3f}')
for n_components in [10, 25, 50, 75, 100]:
    ica = FastICA(n_components=n_components, whiten='arbitrary-variance', max_iter=1000)
    ica.fit(X_train)
    X_train_ica = ica.transform(X_train)
    X_test_ica = ica.transform(X_test)
    clf = SVC(kernel='linear', C=1)
    clf.fit(X_train_ica, y_train)
    y_pred = clf.predict(X_test_ica)
    accuracy = accuracy_score(y_test, y_pred)
    print(f'n_components = {n_components}, ICA accuracy: {accuracy:.3f}')
```
